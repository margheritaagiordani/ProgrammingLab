### **1. Trasformare cicli in list comprehension**

Fai tre esempi di cicli producono una lista e poi riscrivi lo stesso codice usando la list comprehension

```python
#ciclo for
lista1 = []
for i in range(10):
    lista1.append(i)
print(lista1)

#ciclo while
lista3 = []
i = 0
while(i<10):
    lista3.append(i)
    i+=1
print(lista3)

#list comprehension
lista2=[x for x in range(10)]
print(lista2)
```
